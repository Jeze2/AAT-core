```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json

```

```python
right_df = pd.read_csv('superhero_info - superhero_info.csv')
right_df.head()
```

```python
right_df.info()
```

```python
## adding expand=True
right_df['Hero|Publisher'].str.split('|',expand=True)

```

```python
## save the 2 new columns into the dataframe
right_df[['Hero','Publisher']] = right_df['Hero|Publisher'].str.split('|',expand=True)
right_df.head(2)
```

```python
right_df = right_df.drop(columns=['Hero|Publisher'])
right_df
```

```python
## examining a single value from the coordinates col
coord = right_df.loc[0,"Measurements"]
print(type(coord))
coord
```

```python
coord = coord.replace("'",'"')
coord

```

```python
## now we can use json.loads
fixed_coord = json.loads(coord)
print(type(fixed_coord))
fixed_coord
```

```python
## use .str.replace to replace all single quotes
right_df['Measurements'] = right_df['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
right_df['Measurements'] = right_df['Measurements'].apply(json.loads)
right_df['Measurements'].head()

```

```python
lat_long = right_df['Measurements'].apply(pd.Series)
lat_long

```

```python
# concat long_lat with original dataframe
right_df = pd.concat((right_df, lat_long), axis = 1)
right_df.head(2)
```

```python
right_df = right_df.drop(columns=['Measurements'])
right_df
```

```python
right_df['Height'] = pd.to_numeric(right_df['Height'].str.strip(' cm'), errors='coerce')
```

```python
right_df['Weight'] = pd.to_numeric(right_df['Weight'].str.strip(' kg'), errors='coerce')
```

```python
right_df.info()
```

```python
left_df = pd.read_csv('superhero_powers - superhero_powers.csv')
left_df.head()
```

```python
left_df.rename(columns={'hero_names': 'Hero'}, inplace=True)
left_df
```

```python
#Checking data
left_df['Powers'].value_counts()

```

```python
left_df['Powers_2'] = left_df['Powers'].str.split(',')
```

```python
## exploding the column of lists
exploded = left_df.explode('Powers_2')
exploded.head()

```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['Powers_2'].dropna().unique()
cols_to_make

```

```python
for col in cols_to_make:
    left_df[col] = left_df['Powers'].str.contains(col)
left_df.head()

```

```python
## Merging with the same column name
merged = pd.merge(left_df,right_df, on='Hero')
merged
```

```python
# 1. Compare average weight for super speed vs non-super speed
speed_wt = merged[merged['Super Speed']==1]['Weight'].mean()
no_speed_wt = merged[merged['Super Speed']==0]['Weight'].mean()
print(f"Avg weight with super speed: {speed_wt:.2f}") 
print(f"Avg weight without super speed: {no_speed_wt:.2f}")
```

```python
# 2. Average height by publisher
print(merged.groupby('Publisher')['Height'].mean())
```
